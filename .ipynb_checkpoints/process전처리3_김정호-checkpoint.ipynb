```python
maket = pd.read_excel("data/marketing.xlsx",index_col ="기업명",thousands = ",") 
maket.head(16)

```

```python
############### 빈 컬럼제거

maket.info()

del maket["Unnamed: 20"]
```

```python
maket.info()     
```

```python
# 결측치 처리

maketT =maket.T

maketT["그라비티"] = maketT["그라비티"].fillna(maketT.loc[:, "그라비티"].mean())
maketT
```

```python

maket = maketT.T
```

```python

#기업 별 합계 

maket.loc[:,"기업 별 합계"] = maket.sum(axis = 1)  

```

```python
maket
```

```python
maket.sort_values(by="기업 별 합계", ascending=False).head(16)
```

```python
#기업 별 합계 그래프


maket["기업 별 합계"].sort_values().plot(kind="barh", figsize=(15,6),grid=True) 
```

```python
#기업 별 합계 그래프

maket["기업 별 합계"].sort_values().plot(kind="line", figsize=(10,5),grid=True) 
```
