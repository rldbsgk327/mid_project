```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
plt.rc("font", family=font_name)
import datetime
```

```python
corona = pd.read_csv("data/corona.csv")
corona[:] # 전체 데이터 확인
```

```python
corona_descending = corona.sort_values(by='Unnamed: 0' ,ascending=False) #오름차순을 내림차순으로 변경

# 필요 없는 데이터 삭제
del corona_descending["accdefrate"], corona_descending["accexamcnt"], corona_descending["accexamcompcnt"], corona_descending["carecnt"], corona_descending["clearcnt"], corona_descending["deathcnt"], corona_descending["examcnt"], corona_descending["resutlnegcnt"],  corona_descending["seq"], corona_descending["statedt"], corona_descending["statetime"], corona_descending["updatedt"]

corona_descending.columns = ["순서","날짜","누적확진자"] # 컬럼명 지정
corona_descending
```

```python
corona_descending["날짜_"] = pd.to_datetime(corona_descending["날짜"],format='%Y-%m-%d') #날짜를 문자 형식에서 날짜형식으로 변경
del corona_descending["날짜"] # 그 후 이전 데이터(문자형식의 날짜) 삭제
corona_descending
```

```python
plt.plot(corona_descending["날짜_"], corona_descending["누적확진자"], label = "누적확진자(명)")
plt.xlabel("날짜")
plt.ylabel("명",rotation=360)
plt.legend(fontsize=12)
```

```python
df = corona_descending
```

```python
day = df["누적확진자"].values
```

```python
df["일일확진자"] = np.append(day[1:] - day[:-1], 0)
```

```python
df[:]
```

```python
plt.plot(df["누적확진자"], df["일일확진자"]) # 데이터 확인, 마지막 데이터에 이상치 있는 것 확인
```

```python
df = df[:-1] # 마지막 부분 이상치 제거
```

```python
plt.plot(df["누적확진자"], df["일일확진자"]) # 재확인
```

```python
data1Q20 = df[(df['날짜_']>='2020-01-01') & (df['날짜_']<='2020-03-31')] # 20년도 1분기 데이터만 추출
data1Q20 = data1Q20.loc[:, ["일일확진자"]] # 20년도 1분기 데이터 중 일일확진자 데이터 추출
data1Q20.sum() # 전국 20년도 1분기의 평균 기온
df1 = pd.DataFrame(data1Q20.sum())
df1.columns = ["1Q2020"]
df1
```

```python
data2Q20 = df[(df['날짜_'] >=pd.Timestamp(2020, 4, 1)) & (df['날짜_'] <=pd.Timestamp(2020, 6, 30))] # 20년도 2분기 데이터만 추출
data2Q20 = data2Q20.loc[:, ["일일확진자"]] # 20년도 2분기 데이터 중 일일확진자 데이터 추출
data2Q20.sum() # 전국 20년도 2분기 확진자 합
df2 = pd.DataFrame(data2Q20.sum())
df2.columns = ["2Q2020"]
df2
```

```python
data3Q20 = df[(df['날짜_'] >=pd.Timestamp(2020, 7, 1)) & (df['날짜_'] <=pd.Timestamp(2020, 9, 30))] # 20년도 3분기 데이터만 추출
data3Q20 = data3Q20.loc[:, ["일일확진자"]] # 20년도 3분기 데이터 중 일일확진자 데이터 추출
data3Q20.sum() # 전국 20년도 3분기 확진자 합
df3 = pd.DataFrame(data3Q20.sum())
df3.columns = ["3Q2020"]
df3
```

```python
data4Q20 = df[(df['날짜_'] >=pd.Timestamp(2020, 10, 1)) & (df['날짜_'] <=pd.Timestamp(2020, 12, 31))] # 20년도 2분기 데이터만 추출
data4Q20 = data4Q20.loc[:, ["일일확진자"]] # 20년도 4분기 데이터 중 일일확진자 데이터 추출
data4Q20.sum() # 전국 20년도 4분기 확진자 합
df4 = pd.DataFrame(data4Q20.sum())
df4.columns = ["4Q2020"]
df4
```

```python
data1Q21 = df[(df['날짜_'] >=pd.Timestamp(2021, 1, 1)) & (df['날짜_'] <=pd.Timestamp(2021, 3, 31))] # 21년도 1분기 데이터만 추출
data1Q21 = data1Q21.loc[:, ["일일확진자"]] # 21년도 1분기 데이터 중 일일확진자 데이터 추출
data1Q21.sum() # 전국 21년도 1분기 확진자 합
df5 = pd.DataFrame(data1Q21.sum())
df5.columns = ["1Q2021"]
df5
```

```python
data2Q21 = df[(df['날짜_'] >=pd.Timestamp(2021, 4, 1)) & (df['날짜_'] <=pd.Timestamp(2021, 6, 30))] # 21년도 2분기 데이터만 추출
data2Q21 = data2Q21.loc[:, ["일일확진자"]] # 21년도 2분기 데이터 중 일일확진자 데이터 추출
data2Q21.sum() # 전국 21년도 2분기 확진자 합
df6 = pd.DataFrame(data2Q21.sum())
df6.columns = ["2Q2021"]
df6
```

```python
data3Q21 = df[(df['날짜_'] >=pd.Timestamp(2021, 7, 1)) & (df['날짜_'] <=pd.Timestamp(2021, 9, 30))] # 21년도 3분기 데이터만 추출
data3Q21 = data3Q21.loc[:, ["일일확진자"]] # 21년도 3분기 데이터 중 일일확진자 데이터 추출
data3Q21.sum() # 전국 21년도 3분기 확진자 합
df7 = pd.DataFrame(data3Q21.sum())
df7.columns = ["3Q2021"]
df7
```

```python
df_list = [df1.T, df2.T, df3.T,df4.T, df5.T, df6.T,df7.T]
final = pd.concat(df_list)
```

```python
final.plot()
plt.title("분기별 확진자 합")
plt.xlabel("분기")
plt.ylabel("확진자")
plt.show()
```
