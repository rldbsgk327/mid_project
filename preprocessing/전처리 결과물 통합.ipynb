```python
##### 3사매출- 모바일

sales3C_mobile = pd.read_csv("checkpoint/sales_3C_mobile_sum_checkpoint.csv", index_col = 0)
sales3C_mobile
sales3C_mobile.columns = ["3사 Mobile 매출"]
sales3C_mobile
```

```python
##### 3사매출- PC

sales3C_pc = pd.read_csv("checkpoint/sales_3C_pc_sum_checkpoint.csv", index_col = 0)
sales3C_pc

sales3C_pc.columns = ["3사 PC매출"]
sales3C_pc
```

```python
##### 날씨

temperature = pd.read_csv("checkpoint/temperature_checkpoint.csv", index_col = 0)
temperature = temperature.T
temperature
```

```python
###### 마케팅 비용

marketing_cost = pd.read_csv("checkpoint/marketing_cost_checkpoint.csv", index_col = 0)
marketing_cost = marketing_cost.T
marketing_cost = marketing_cost.drop(index=["기업 별 합계"], axis=0)

marketing_cost_sum = marketing_cost.loc[:, "분기별 합계"]
marketing_cost_sum
```

```python
marketing_cost
```

```python
##### 게임시장 매출

sales_all = pd.read_csv("checkpoint/gamesales_all_checkpoint.csv",index_col = 0)

sales_all_sum = sales_all.T.loc[:, "Sum"]
sales_all_sum
```

```python
# 통합

result = pd.concat([sales3C_mobile,sales3C_pc, temperature, marketing_cost_sum, sales_all_sum],axis=1)
result
```

```python
result.rename(columns={'분기별 합계':'마케팅 비용'}, inplace = True)
result.rename(columns={'Sum':'게임시장 규모'}, inplace = True)
```
